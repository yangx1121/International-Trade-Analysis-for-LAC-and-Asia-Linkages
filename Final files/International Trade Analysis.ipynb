```python
%%sql
WITH CHN AS(
    SELECT year, partneriso3, SUM(value_in_000) AS sum1
    FROM trade
    WHERE partneriso3 = 'CHN' AND tradeflow='Import'
    GROUP by year, partneriso3), 
WLD AS(
    SELECT year, partneriso3, SUM(value_in_000) AS sum2
    FROM trade
    WHERE partneriso3 = 'WLD' AND tradeflow='Import'
    GROUP by year, partneriso3)
SELECT wld.year, 100*sum1/sum2 AS perc FROM CHN, WLD WHERE CHN.year=WLD.year order by year
```

```python
%%sql

WITH cte AS (
    SELECT year, partneriso3, sum(value_in_000) AS total
    FROM trade
    WHERE partneriso3 = 'WLD' or partneriso3 = 'CHN' AND tradeflow = 'Import'
    GROUP BY year, partneriso3
    )

select year, partneriso3, total, sum(total) over(partition by year) as sum1, total/sum(total) over(partition by year)
from cte
```

### Category x Country

```python
%%sql
WITH CHN AS(
    SELECT trade.year, trade.partneriso3, SUM(trade.value_in_000) AS sum1
    FROM trade 
    JOIN product
    ON trade.productcode = product.productcode
    AND trade.partneriso3 = 'CHN' 
    AND trade.tradeflow='Import' 
    AND product.h2_category='Vegetable'
    GROUP by trade.year, trade.partneriso3), 
WLD AS(
    SELECT trade.year, trade.partneriso3, SUM(trade.value_in_000) AS sum2
    FROM trade
    JOIN product
    ON trade.productcode = product.productcode
    AND trade.partneriso3 = 'WLD' 
    AND trade.tradeflow='Import' 
    AND product.h2_category='Vegetable'
    GROUP by trade.year, trade.partneriso3)
SELECT wld.year, 100*sum1/sum2 AS perc FROM CHN, WLD WHERE CHN.year=WLD.year ORDER BY year
```
